1. Write a Python program to read a CSV file and display its contents.

First way

```python
import csv #import library

# Open and read the CSV file
with open('/Users/jitendragupta/Downloads/Housing.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    
    # Read the header
    header = next(csv_reader)
    print('Header:', header)
    
    # Read and print each row
    for row in csv_reader:
        print(row)
```

2. Write a Python function to reverse a string.

```python
#define a function reverse_string()
def reverse_string(s):
    return s[::-1]

original_string = "hello world!"
reversed_string = reverse_string(original_string)
print(reversed_string) #print the output
```

3. Write a Python function to count the occurrences of a specific element in a list.

```python
#define a function count_occurence()
def count_occurrences(lst, element):
    return lst.count(element)


List1 = [1, 2, 3, 1, 2, 1, 4, 1, 8, 3, 9, 1] #define a list

print(count_occurrences(List1, 1)) #print the count for a element - 1
```
